```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Datasets/WineQT.csv')

# converting quality attribute
df['quality'] = df['quality'].apply(lambda x: 0 if x <= 6 else 1)

# Normalizing other attributes
scaler = MinMaxScaler()
features = df.drop('quality', axis=1)
scaled_features = scaler.fit_transform(features)

# answer dataframe
ans_df = pd.DataFrame(scaled_features, columns=features.columns)
ans_df['quality'] = df['quality']

# answer dataset
ans_df.to_csv('scaled_winequality.csv', index=False)
```
